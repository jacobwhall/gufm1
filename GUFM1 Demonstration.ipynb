```python
import numpy as np
import matplotlib.pyplot as plt
import gufm1
%matplotlib inline
```

```python
year = 1990
g,h = gufm1.get_gh_at_t(year)
```

```python
r = 3480.
th = np.linspace(0,np.pi,100)
ph = -10*np.pi/180.
Br1990 = gufm1.Br(r,th,ph,g,h)
```

```python
plt.plot(th*180/np.pi, Br1990/1e6)
plt.grid()
plt.title('Radial Magnetic Field Strength at CMB\nL_max = 14, -10deg W, 1990')
plt.xlabel('colatitude')
plt.ylabel('magnetic field strength (mT)')
```
